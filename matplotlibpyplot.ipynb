## Bassic Graph

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

## Resizing my graph
# plt.figure(figsize=(2.5,1.5), dpi = 300)

plt.plot(x,y, label = '2x', color='green', linestyle='--' , linewidth=2, marker = '*' , markersize=5 , markeredgecolor='gray')

plt.title('Population graph', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20} )
plt.xlabel('X - axis')
plt.ylabel('Y - axis')

## Select interval we want to plot pints at
x2 = np.arange(0,4.5,0.5)
# plt.plot(x2, x2**2, 'r', label='x^2')

plt.plot(x2[:6], x2[:6]**2, 'r', label='x^2')
plt.plot(x2[5:],x2[5:]**2, 'r--')

plt.xticks([0,1,2,3,4])
# plt.yticks([0,2,4,6,8])

plt.legend() 

plt.savefig('myLineGraph.png', dpi=300)

plt.show()
```

# Bar Charts


```python
labels = ['A', 'B', 'C']
values = [1,4,2]

bars = plt.bar(labels, values, label="Life span")

plt.title("Population Bar Graph")

plt.xlabel("Age group")
plt.ylabel("frequecy density")

pattern = ['/', '0', '*']
for bar in bars:
    bar.set_hatch(pattern.pop(0))

# bars[0].set_hatch('/')
# bars[1].set_hatch('o')
# bars[2].set_hatch('*')

plt.savefig('myBarGraph.png', dpi = 300)

plt.legend()

plt.show()
```

```python
N = 5
menMeans = (20, 35, 30, 35, -27)
womenMeans = (25, 32, 34, 20, -25)
menStd = (2, 3, 4, 1, 2)
womenStd = (3, 5, 2, 3, 3)
ind = np.arange(N)    # the x locations for the groups
width = 0.35       # the width of the bars: can also be len(x) sequence

fig, ax = plt.subplots()

p1 = ax.bar(ind, menMeans, width, yerr=menStd, label='Men')
p2 = ax.bar(ind, womenMeans, width,
            bottom=menMeans, yerr=womenStd, label='Women')

ax.axhline(0, color='grey', linewidth=0.8)
ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.set_xticks(ind, labels=['G1', 'G2', 'G3', 'G4', 'G5'])
ax.legend()

# Label with label_type 'center' instead of the default 'edge'
ax.bar_label(p1, label_type='center')
ax.bar_label(p2, label_type='center')
ax.bar_label(p2)

plt.show()
```
